```python
def normalize_json(dataframe):
  df_return=pd.DataFrame()
  for column in list(df.columns[2:]):
    df_column=pd.json_normalize(df[column], sep='')
    df_return=pd.concat([df_return,df_column], axis=1)
  return pd.concat([dataframe[list(df.columns[:2])], df_return], axis=1)
```

```python
df_norm=normalize_json(df)
df_norm
```
